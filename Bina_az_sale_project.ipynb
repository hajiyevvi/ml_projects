```python
# bu kod vasitesi ile sahə sütunundakı sotla verilmis ededleri m2 ceviririk(1sot = 100 m2)

import re
def convert_to_m2(x):
    # Boşluqları təmizləyirik
    x = x.strip()
    # Rəqəm və vahidi ayırmaq üçün regex istifadə edirik
    match = re.match(r'([\d\.]+)\s*(m²|sot)', x)
    if match:
        number = float(match.group(1))
        unit = match.group(2)
        if unit == 'm²':
            return number
        elif unit == 'sot':
            return number * 100  # 1 sot = 100 m²
    return None  # Əgər uyğunluq tapılmazsa
df['Sahə'] = df['Sahə'].apply(convert_to_m2)
```

#Base Model

```python
num_pipeline_log = make_pipeline(
    SimpleImputer(strategy = 'median'),
    StandardScaler(),
    PolynomialFeatures(degree = 2)
)

cat_pipeline_log = make_pipeline(
    SimpleImputer(strategy = 'most_frequent'),
    OneHotEncoder(handle_unknown= 'ignore'),

)

preprocessing_log = ColumnTransformer(
    transformers = [
        ('num', num_pipeline_log, num_columns),
        ('cat',cat_pipeline_log, cat_columns)
    ]
)

log_reg = LogisticRegression(class_weight = {0: 0.65, 1: 0.35})

full_pipeline_log = make_pipeline(preprocessing_log, log_reg)
full_pipeline_log.fit(X_train, y_train)
```

```python
full_pipeline_log.score(X_test,y_test)
```

```python
full_pipeline_log.score(X_train,y_train)
```

```python
y_pred = full_pipeline_log.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# inbalance datasetlerde f1 score daha önəmlidir.Bunun üçün f1 score nəzərə alırıq.

f1 = f1_score(y_test,y_pred,pos_label='var',average = 'weighted')
f1
```

# DecisionTreeClassification

```python
num_pipeline_dec  = make_pipeline(
    SimpleImputer(strategy = 'median')
)

cat_pipeline_dec = make_pipeline(
    SimpleImputer(strategy = 'most_frequent'),
    OneHotEncoder(handle_unknown= 'ignore')

)

preprocessing_dec = ColumnTransformer(
    transformers = [
        ('num', num_pipeline_dec, num_columns),
        ('cat',cat_pipeline_dec, cat_columns)
    ],sparse_threshold=0.0
)

dec_tree = DecisionTreeClassifier(class_weight = {0: 0.65, 1: 0.35},min_samples_split = 4, min_samples_leaf = 3, max_depth = 46)

full_pipeline_dec = make_pipeline(preprocessing_dec, dec_tree)
full_pipeline_dec.fit(X_train, y_train)
```

```python
full_pipeline_dec.score(X_test,y_test)
```

```python
full_pipeline_dec.score(X_train, y_train)
```
